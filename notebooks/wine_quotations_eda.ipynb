```python
import plotly.graph_objs as go


fig = go.Figure()


# for row in wine_quotations_df.field_vintage_value:
for name,group in  wine_quotations_df.groupby(["name","field_vintage_value"]):
    fig.add_trace(go.Scatter(x=group.field_date_value, y=group.field_price_amount,
#                              name = str(name),
                             mode = 'lines')
                 )

# one button for each df column
updatemenu= []
buttons=[]

# for row in wine_quotations_df.name.unique():
#     buttons.append(dict(method='restyle',
#                         label=row,
#                         args=[{'y':[wine_quotations_df.field_price_amount]}])
#                   )
    
for name,group in  wine_quotations_df.groupby(["name","field_vintage_value"]):
    buttons.append(dict(method='restyle',
                        label=str(name),
                        args=[{'y':[group.field_price_amount]}])
                  )   
# some adjustments to the updatemenu
updatemenu=[]
your_menu=dict()
updatemenu.append(your_menu)
updatemenu[0]['buttons']=buttons
updatemenu[0]['direction']='down'
updatemenu[0]['showactive']=True

# update layout and show figure
fig.update_layout(updatemenus=updatemenu)
fig.show()
```
